# Error using train test split

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(car)
```

```python
car_std = scaler.transform(car)
```

```python
car1 = pd.DataFrame(car_std, columns=car.columns)
```

```python
X = car1.drop(columns = "MSRP")
y = car1['MSRP']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np


```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)

# Evaluate the KNN model
y_pred_knn = knn.predict(X_test)
mae_knn = mean_absolute_error(y_test, y_pred_knn)
mse_knn = mean_squared_error(y_test, y_pred_knn)
rmse_knn = np.sqrt(mse_knn)
r2_knn = r2_score(y_test, y_pred_knn)
acc_knn = knn.score(X_train, y_train)

print('KNN - Training accuracy:', acc_knn)
print('KNN - Test accuracy:', knn.score(X_test, y_test))
print("KNN - MAE:", mae_knn)
print("KNN - MSE:", mse_knn)
print("KNN - RMSE:", rmse_knn)
print("KNN - R-squared:", r2_knn)
```

```python
import matplotlib.pyplot as plt

# Plot the evaluation metrics
labels = ['MAE', 'MSE', 'RMSE']
values = [mae_knn, mse_knn, rmse_knn]
plt.bar(labels, values)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=22)
rf.fit(X_train, y_train)

# Evaluate the Random Forest model
y_pred_rf = rf.predict(X_test)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, y_pred_rf)
acc_rf = rf.score(X_train, y_train)

print('Random Forest - Training accuracy:', acc_rf)
print('Random Forest - Test accuracy:', rf.score(X_test, y_test))
print("Random Forest - MAE:", mae_rf)
print("Random Forest - MSE:", mse_rf)
print("Random Forest - RMSE:", rmse_rf)
print("Random Forest - R-squared:", r2_rf)
```
